```python
from bs4 import BeautifulSoup
import urllib
import pandas as pd
```

```python
url = "https://xeno-canto.org/explore?query=brazil"
url = "https://xeno-canto.org/explore?query=+type%3A%22call%22+cnt%3A%22%3DBrazil%22"
pages = []
for pg in range(1,793):
    print("Page:", pg)
    with urllib.request.urlopen(url + "&pg="+str(pg)) as response:
        html = response.read().decode()
        pages.append(html)
```

```python
print(len(pages))
```

```python
sci_names = []
com_names = []
links = []
cnt = 1
for html in pages:
    print("Page:", cnt)
    cnt += 1
    soup = BeautifulSoup(html, 'html.parser')
    res = soup.find_all("tr")[2:-1]

    for row in res:
        scientific_name = row.find(class_="scientific-name").get_text()
        common_name = row.find(class_="common-name").get_text()
        button_audio = row.find(class_="xc-button-audio")
        if button_audio is not None:
            link = button_audio.find("div")['data-xc-filepath'][2:]
            link = "http://www." + link
            sci_names.append(scientific_name)
            com_names.append(common_name)
            links.append(link)
```

```python
data = {'scientific name': sci_names, 'common name': com_names, 'audiofile': links}
df = pd.DataFrame(data)
df.head()
```

```python
print(len(df))
meta = df.groupby(['common name']).count().sort_values(['scientific name'], ascending=False)
```

```python
df.to_csv('./brazil_birds_at_xenocanto.csv')
```
