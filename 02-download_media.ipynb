```python
import pandas as pd
import urllib
import os
```

```python
df = pd.read_csv('brazil_birds_at_xenocanto.csv')
df.head()
```

```python
df_ = df.groupby('common name').count()
filter = df_['audiofile'] >= 50
df_ = df_[filter].reset_index()
print(len(df_))
names = set(df_['common name'])
cn = list(df['common name'])
filter2 = [c in names for c in cn]
df = df[filter2]
print(len(df))
```

```python
names = list(df['common name'])
links = list(df['audiofile'])
local_files = list()

for n in range(len(names)):
    url = links[n]
    data = urllib.request.urlopen(url)
    local_fname = names[n].replace(' ', '_').replace('\n', '') + '-' + url.split('/')[-1][-50:] # Windows doesnt like very long filenames
    local_fname = local_fname.lower()
    local_files.append(local_fname)
    print(n, local_fname)
    f = open('audio/'+local_fname,'wb')
    f.write(data.read())
    f.close()
```

```python
df['local filename']=local_files
```
